```python
import cv2
import os
img_folder_path = "C:\\Users\\8902\\Desktop\\102flowers\\jpg"
output_folder_path = "C:\\Users\\8902\\Desktop\\102flowers\\resizeimage"
# define the size of the images after resize
IMG_SIZE = (256, 256)
# define the size of the images after resize
IMG_SIZE = (256, 256)

# loop through each image in the folder
for img_name in os.listdir(img_folder_path):
    # read the image
    img_path = os.path.join(img_folder_path, img_name)
    img = cv2.imread(img_path)

    # resize the image
    resized_img = cv2.resize(img, IMG_SIZE)

    # save the resized image
    output_path = os.path.join(output_folder_path, img_name)
    cv2.imwrite(output_path, resized_img)
```

```python
from PIL import Image
import numpy as np
import os

# directory path where the resized images are saved
dir_path =  "C:\\Users\\8902\\Desktop\\102flowers\\resizeimage"

# create a new directory to save the normalized images
if not os.path.exists(os.path.join(dir_path, "normalized")):
    os.makedirs(os.path.join(dir_path, "normalized"))

# loop through all images in the directory
for filename in os.listdir(dir_path):
    if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
        # open image and convert to numpy array
        img_path = os.path.join(dir_path, filename)
        img = Image.open(img_path)
        img_arr = np.array(img)

        # normalize pixel values to range [0, 1]
        img_normalized = img_arr / 255.0

        # convert back to PIL image
        img_normalized_pil = Image.fromarray(np.uint8(img_normalized * 255))

        # save normalized image
        img_normalized_pil.save(os.path.join(dir_path, "normalized", filename))
```
